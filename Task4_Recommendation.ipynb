```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

# Sample user-movie rating dataset
data = {
    'Movie1': [5, 4, 0, 0],
    'Movie2': [4, 0, 0, 2],
    'Movie3': [0, 0, 5, 4],
    'Movie4': [0, 3, 4, 5]
}

df = pd.DataFrame(data, index=['User1', 'User2', 'User3', 'User4'])

print("User-Movie Ratings:\n")
print(df)

# Compute similarity
similarity = cosine_similarity(df)

similarity_df = pd.DataFrame(similarity, index=df.index, columns=df.index)

print("\nUser Similarity Matrix:\n")
print(similarity_df)

# Plot heatmap
plt.imshow(similarity)
plt.title("User Similarity Matrix")
plt.colorbar()
plt.xticks(range(len(df.index)), df.index)
plt.yticks(range(len(df.index)), df.index)
plt.show()
```
